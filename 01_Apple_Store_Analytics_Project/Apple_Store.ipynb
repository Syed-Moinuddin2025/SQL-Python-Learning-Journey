### 📌 Step 2: Load All CSV Files


```python
df_stores = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\stores.csv")
df_products = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\products.csv")
df_category = pd.read_csv(r'D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\category.csv')
df_sales = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\sales.csv")
df_warranty = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\warranty.csv")
```

### 📌 Step 4: Merge Tables


#### Merge products with category


```python
products_full = pd.merge(df_products, df_category, on='category_id', how='left')
```

#### Merge sales with products_full


```python
sales_products = pd.merge(df_sales, products_full, on='product_id', how='left')
```

#### Merge with stores


```python
data_merged = pd.merge(sales_products, df_stores, on='store_id', how='left')
```

#### 📌 Merge warranty into sales first


```python
sales_with_warranty = pd.merge(df_sales, df_warranty, on='sale_id', how='left')
```

#### 📌 Merge with products + category


```python
sales_products = pd.merge(sales_with_warranty, products_full, on='product_id', how='left')
```

#### 📌 Merge with stores


```python
data_final = pd.merge(sales_products, df_stores, on='store_id', how='left')
```

### 📊 Sales Analysis

```python
# 1. What is the total revenue generated across all stores?
total_revenue = data_final['price'].sum()
print(f"\nTotal Revenue Generated Across All Stores: ${total_revenue:,.2f}")
```

```python
# 2. Which product generated the highest revenue?
highest_revenue_product = data_final.groupby('product_name')['price'].sum().idxmax()
highest_revenue_value = data_final.groupby('product_name')['price'].sum().max() 
print(f"\nProduct with Highest Revenue: {highest_revenue_product} (${highest_revenue_value:,.2f})")
```

```python
#3. What is the total revenue by product category?
# Create revenue column first
data_final['revenue'] = data_final['quantity'] * data_final['price']

# Group by category and sum revenue
total_revenue_by_category = (
    data_final.groupby('category_name')['revenue']
    .sum()
    .reset_index()
    .sort_values(by='revenue', ascending=False)
)

# Show result
print(total_revenue_by_category)
```

```python
# 4. Which store has the highest number of sales?
# Total sales quantity per store
store_sales = data_final.groupby('store_name')['quantity'].sum().reset_index()

# Sort in descending order
store_sales = store_sales.sort_values(by='quantity', ascending=False)

# Top store
top_store = store_sales.iloc[0]

# Show result
print("🏬 Highest Sales Store:")
print(f"{top_store['store_name']} → {top_store['quantity']} units sold")
```

```python
#5. Which city has the highest total revenue?
city_revenue = data_final.groupby('city')['revenue'].sum().reset_index()
city_revenue = city_revenue.sort_values(by='revenue', ascending=False)  
top_city = city_revenue.iloc[0]
print(f"\n🏙️ City with Highest Revenue: {top_city['city']} → ${top_city['revenue']:,.2f}")
```

```python
#6. What is the average sale value per store?
average_sale_per_store = data_final.groupby('store_name')['revenue'].mean().reset_index()
average_sale_per_store = average_sale_per_store.sort_values(by='revenue', ascending=False)
print("\n🔝 Top 5 Stores by Avg Sale Value:")
print(average_sale_per_store.head())
```

#### 📦 Product Performance

```python
#1. What are the top 5 best-selling products by quantity?
top_products_by_quantity = (
    data_final.groupby('product_name')['quantity']
    .sum()
    .reset_index()
    .sort_values(by='quantity', ascending=False)
)
print("\n📦 Top 5 Best-Selling Products by Quantity:")
print(top_products_by_quantity.head())
```
