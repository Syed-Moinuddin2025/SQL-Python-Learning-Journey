### 📌 Step 2: Load All CSV Files


```python
df_stores = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\stores.csv")
df_products = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\products.csv")
df_category = pd.read_csv(r'D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\category.csv')
df_sales = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\sales.csv")
df_warranty = pd.read_csv(r"D:\sql_python_mastery\Python_Journey\01_Apple_Store_Analytics_Project\warranty.csv")
```

### 📌 Step 3: Preview Basic Info


```python
print("\n✅ Stores:")
print(df_stores.head())
print("\n✅ Products:")
print(df_products.head())
print("\n✅ Categories:")
print(df_category.head())
print("\n✅ Sales:")
print(df_sales.head())
print("\n✅ Warranty:")
print(df_warranty.head())
```

### 📌 Step 4: Merge Tables


#### Merge products with category


```python
products_full = pd.merge(df_products, df_category, on='category_id', how='left')
```

#### Merge sales with products_full


```python
sales_products = pd.merge(df_sales, products_full, on='product_id', how='left')
```

#### Merge with stores


```python
data_merged = pd.merge(sales_products, df_stores, on='store_id', how='left')
```

#### 📌 Merge warranty into sales first


```python
sales_with_warranty = pd.merge(df_sales, df_warranty, on='sale_id', how='left')
```

#### 📌 Merge with products + category


```python
sales_products = pd.merge(sales_with_warranty, products_full, on='product_id', how='left')
```

#### 📌 Merge with stores


```python
data_final = pd.merge(sales_products, df_stores, on='store_id', how='left')
```
